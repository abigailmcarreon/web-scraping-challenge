```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import os 
import requests
import pymongo
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# URL of NASA Mars News
url = 'https://redplanetscience.com'
```

```python
#with browser as browser:
browser.visit(url)
mars_html = browser.html
mars_soup = bs(mars_html, 'html.parser')
```

```python
print(mars_soup.prettify())
```

```python
news_title = mars_soup.find(class_='content_title').text
print(news_title)
```

```python
news_paragraph = mars_soup.find(class_='article_teaser_body').text
print(news_paragraph)
```

```python
#Set URL for JPL Mars Space Images - Featured Image
spaceimages_url = 'https://spaceimages-mars.com'
```

```python
#with browser as browser:
browser = Browser('chrome', **executable_path, headless=True)
browser.visit(spaceimages_url)
spaceimages_html = browser.html
spaceimages_soup = bs(spaceimages_html, 'html.parser')
```

```python
#Image URL for JPL Mars Space Images - Featured Image
#Not full URL like in example, but oh well
featured_image_url = spaceimages_soup.find('img', class_='headerimage')['src']
print(featured_image_url)
```
